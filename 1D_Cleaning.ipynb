```python
import pandas as pd
```

```python
df = pd.read_excel('data/table_1d_referrals_completing_treatment_ccg_problem_descriptor_2014_15.xlsx')
df.head(250)
```

```python
# Drop rows that are completely empty
df = df.dropna(how="all")

# Drop columns that are completely empty
df = df.dropna(axis=1, how="all")

# Rename the third column to 'CCG Name'
df = df.rename(columns={df.columns[2]: 'CCG Name'})

# Merge first two columns together and rename to 'CCG Code'
df['CCG Code'] = df.iloc[:, 0].fillna('') + ' ' + df.iloc[:, 1].fillna('')
df = df.drop(df.columns[[0, 1]], axis=1)

df.head(250)
```

```python
#Merge rows 4 and 5 to create proper header and remove 'of which' in Unnamed: 7 
#Make sure to leave CCG Name and CCG Code alone
new_header = df.iloc[4].fillna('') + ' ' + df.iloc[5].fillna('') 
new_header = new_header.str.replace(' of which', '', regex=False).str.strip() 
df = df[6:] 
df.columns = new_header 
df.head(250)
```

```python
#Remove row with England total
df = df[df['CCG Code'].str.strip() != 'England Total']

#Drop Invalid Data Supplied column if exists
if 'Invalid data supplied' in df.columns:
    df = df.drop(columns=['Invalid data supplied'])
df.head(250)
```
