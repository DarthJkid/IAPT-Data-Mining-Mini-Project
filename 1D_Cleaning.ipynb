```python
import pandas as pd
```

```python
df = pd.read_excel('data/table_1d_referrals_completing_treatment_ccg_problem_descriptor_2014_15.xlsx')
df.head(250)
```

```python
#Merge Unnamed:2 and CCG Code
df['CCG Code'] = df['Unnamed: 2'].fillna(df['CCG Code'])
df = df.drop(columns=['Unnamed: 2'])
```

```python
df.head(250)
```

```python
#Drop Unnamed: 0 and Invalid data supplied
df = df.drop(columns=['Unnamed: 0', 'Invalid data supplied'])
df.head(250)
```

```python
#Replace missing values with '*'
df = df.fillna('*')
df.head(250)
```

```python
#save df to new excel file
df.to_excel('cleaned_data/cleaned_table_1d.xlsx', index=False)
```
