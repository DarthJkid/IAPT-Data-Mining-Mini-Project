```python
import pandas as pd

file_path = "psych-ther-ann-rep-tab-2014-15.xlsx"

xls = pd.ExcelFile(file_path)
print(xls.sheet_names)
```

```python
raw = pd.read_excel(file_path, sheet_name="Table 1a", header=None)
raw.head(20)   # inspect the first 20 rows
```

```python
header_row = 5# adjust based on what you saw in raw.head(20)

df = pd.read_excel(
    file_path,
    sheet_name="Table 1a",
    header=header_row
)

df.head()
```

```python
# Drop rows that are completely empty
df = df.dropna(how="all")

# Drop columns that are completely empty
df = df.dropna(axis=1, how="all")


df.head()
```

```python
# Pick a numeric column you know should have data
key_col = df.select_dtypes(include=["number"]).columns[0]
df = df[df[key_col].notna()]
```
