```python
header_row = 5
df = pd.read_excel('data/table_1b_referrals_treatment__general_ccg_2014_15.xlsx',
    header=header_row
)

df.head(15)
```

```python
# Drop rows that are completely empty
df = df.dropna(how="all")

# Drop columns that are completely empty
df = df.dropna(axis=1, how="all")

#Drop row 2
df = df.drop(index=2)

#Rename Unnamed: 3 to 'CCG Name'
df = df.rename(columns={df.columns[2]: 'CCG Name'})

#Merge Unnamed:1 and 2 together and rename to 'CCG Code'
df['CCG Code'] = df.iloc[:, 0].fillna('') + ' ' + df.iloc[:, 1].fillna('')
df = df.drop(df.columns[[0, 1]], axis=1)

df.head(15)
```

```python
df.head(250)
```

```python
#Drop Lancashire Commissioning Hub and West Yorkshire Commissioning Hub rows
df = df[~df['CCG Name'].isin(['Lancashire Commissioning Hub', 'West Yorkshire Commissioning Hub'])]
df.head(250)
```

```python
#Drop everything after the first occurrence of 'Data Source' in 'CCG Code' column
unknown_index = df[df['CCG Code'].str.contains('Data source', na=False)].index
if not unknown_index.empty:
    df = df.loc[:unknown_index[0]-1]

#Put 'UNK' in CCG Code for rows where CCG Name is 'Unknown CCG'
df.loc[df['CCG Name'] == 'Unknown CCG', 'CCG Code'] = 'UNK'
df.head(250)
```

```python
#Save cleaned dataframe to new excel file
df.to_excel('cleaned_data/cleaned_table_1b.xlsx', index=False)
```
